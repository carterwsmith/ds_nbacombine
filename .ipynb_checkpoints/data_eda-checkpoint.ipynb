```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df[['STANDING_REACH', 'STANDING_VERTICAL_LEAP', 'BLK', 'DREB', 'HEIGHT_W_SHOES']].corr(), cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
df_stat = df[['MAX_VERTICAL_LEAP', 'THREE_QUARTER_SPRINT', 'PTS']]
for i in df_stat:
    num = df_stat[i].value_counts()
    bar = sns.barplot(x=num.index, y=num)
    bar.set(xlabel=i, ylabel='Number of Players')
    bar.set_xticklabels(bar.get_xticklabels(), rotation=90, horizontalalignment='right')
    plt.show()
```
